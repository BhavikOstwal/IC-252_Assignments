```python
import numpy as np
import matplotlib.pyplot as plt

# Values for the histograms
value_1 = 1.93
value_2 = -0.71
value_3 = -0.02
value_4 = 0.29
value_5 = 0.08

# Create datasets for histogram
data_1 = [value_1]
data_2 = [value_2]
data_3 = [value_3]
data_4 = [value_4]
data_5 = [value_5]

# Combine both datasets
combined_data = [value_1, value_2,value_3,value_4,value_5]

plt.figure(figsize=(8,7))
# Plot histograms side by side
plt.bar(['2017', '2018','2019','2020','2021'],combined_data, color=['green','green','green','red','red'], edgecolor='black', label=['2017', '2018','2019','2020','2021'], alpha=1)
plt.xlabel('year')
plt.ylabel('Avg % change')
plt.title('Stock Investing Returns')
plt.legend()

# Show plot
plt.grid(alpha=0.5)
plt.savefig('../figures/stock_returns.png')
plt.show()

```
