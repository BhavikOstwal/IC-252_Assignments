```python
import numpy as np
import matplotlib.pyplot as plt

X_val = np.arange(0,6)
No_of_students = np.array([2,11,23,9,4,1])
# print(X_val)

n = 50
PDF_val =  No_of_students/n
# print(PDF_val.dot(X_val))

CDF_val = np.cumsum(PDF_val)

plt.title("CDF")
plt.step(X_val, CDF_val, '-o', where='post', color='r')
plt.plot(np.arange(5,7), np.ones((2,)), color='r')
plt.plot(np.arange(-1,1), np.zeros((2,)), color='r')
plt.xlim(-0.5,5.5)
plt.ylim(-0.1,1.1)
plt.grid()
plt.show()
```
