```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def factorial(n:int):
    if (n<2):
        return 1
    return n*factorial(n-1)
```

```python
k_values = []
p_values = []
for k in range(2,51):
    a = factorial(365)
    a/= (factorial(365-k))*(365**k)
    k_values.append(k) 
    p_values.append(1-a) 
```

```python
plt.plot(k_values, p_values)
plt.show()
```

```python
def factroial_s(n:int):
    return ((2*np.pi*n)**0.5)*((n/math.e)**n)
```

```python
n_val = []
ratio_val = []
for i in range(1,21):
    n_val.append(i)
    ratio_val.append(factorial(i)/factroial_s(i))
```

```python
plt.plot(n_val, ratio_val)
plt.ylim(0,2)
plt.show()
```

```python
n_values = np.arange(2,501)
prob_values=[]
for n in range(len(n_values)):
    arr = np.arange(1,n_values[n]+1)
    rlt_frq =0
    for j in range(1000):
        np.random.shuffle(arr)
        for i in range(len(arr)):
            if (arr[i]==i+1):
                rlt_frq+=1
                break
    prob_values.append(rlt_frq/1000)
        
plt.plot(n_values, prob_values)
plt.show()
```
