```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/BBRI.csv')


mean_values = df.mean()
colors = ['g', 'g', 'g', 'r', 'r']

# plt.style.use('bmh')
plt.figure(figsize=(8,7))

mean_values.plot(kind='bar', title='Stock Return of BBRI for 2017-2021',color=colors,edgecolor='black', xlabel='Year', ylabel='Mean Value')
plt.grid(alpha=0.5)
plt.savefig('../figures/BBRI.png')
```
